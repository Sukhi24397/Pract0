```python
'''4.A dictionary stores student names and marks.Write a python program to find the student who scored the highest marks.'''

marks={"A":85,"B":92,"C":78}
topper=max(marks,key=marks.get)
print("topper:",topper)
print("marks:",marks[topper])
```
